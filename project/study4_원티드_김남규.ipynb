```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import pandas as pd 
from urllib.parse import urljoin
```

```python
driver = webdriver.Chrome("")
time.sleep(3)
url_base='https://www.wanted.co.kr'
url_sub='/wdlist/518/163lastpage = int(input("마지막 페이지 번호를 입력하세요."))4?country=kr&job_sort=job.recommend_order&years=-1&selected=1634&locations=seoul.songpa-gu&locations=seoul.gangnam-gu&locations=seoul.guro-gu&locations=seoul.geumcheon-gu&locations=seoul.seocho-gu'
url=url_base + url_sub
time.sleep(3)
driver.get(url)
time.sleep(3)
```

```python
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 스크롤 끝까지 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # 스크롤 내린 후 페이지 로딩을 위한 시간이 필요하다면, PAUSE_SEC에 숫자 입력
    time.sleep(2)
    req=driver.page_source
    soup=BeautifulSoup(req, 'html.parser')
    soup_list=soup.find_all('li',class_='Card_Card__hfQPt')
    len(soup_list)
    
    company_element=soup.find_all('span',class_='Typography_Typography__root__xYuMs Typography_Typography__label2__BRssq Typography_Typography__weightMedium__O0Rdi JobCard_JobCard__body__company__F6XoH')
    company_list = []
    for item in company_element:
        company = item.text
        company_list.append(company)
    
    title_element=soup.find_all('p',class_ = 'Typography_Typography__root__xYuMs Typography_Typography__body1__UmyQW Typography_Typography__weightBold__e15ql JobCard_JobCard__body__position__P8R0W')
    title_list=[]
    for item in title_element:
        title= item.text
        title_list.append(title)
    
    short_add_element=soup.find_all('span',class_='Typography_Typography__root__xYuMs Typography_Typography__label2__BRssq Typography_Typography__weightMedium__O0Rdi JobCard_JobCard__body__location__hz8BQ')   
    short_add_list=[]
    career_list=[]
    for item in short_add_element:
        short_add= item.text
        short=short_add.split('·')[0]
        short_add_list.append(short)
        career= item.text
        career=short_add.split('·')[1]
        career_list.append(career)   
    
    url_element = soup.find_all('div', 'JobCard_JobCard__oZL4d')
    url_add_list=[]
    for item in url_element:
        url_add_list.append(urljoin(url_base, item.find('a')['href']))
    # 스크롤 내린 후의 페이지 높이 = new_height
    new_height = driver.execute_script("return document.body.scrollHeight")

    # 더이상 스크롤이 내려가지 않으면 스크롤 내리는 반복문 멈춤 (=last_height와 new_height 같으면 멈춤)
    if new_height == last_height:
        break
    # 스크롤 내린 후의 페이지 높이(new_height)를 현재 페이지 높이(last_height) 변수에 저장
    last_height = new_height  
    
    
      
    
```

```python
position_list=[]
tec_tag_list=[]
tag_list=[]
dday_list=[]
address_list=[]

total_rows = len(df)

batch_size = 20

for start in range(0, total_rows, batch_size):
    driver = webdriver.Chrome("")
    end = min(start + batch_size, total_rows)
    for idx, row in df[start:end].iterrows():
        url=row['URL']
        driver.get(url)
        time.sleep(0.5)
        soup_tmp=BeautifulSoup(driver.page_source,'html.parser')
        time.sleep(0.5)

        try:
            tec_tag=driver.find_element(By.CLASS_NAME,'JobSkillTags_JobSkillTags__list__eEc2V').text.split('\n')
        except:
            tec_tag=None
        try:
            tag=driver.find_element(By.CLASS_NAME,'CompanyTags_CompanyTags__list__KgTBZ').text.split('\n')
        except:                                 
            tag=driver.find_element(By.XPATH ,'//*[@id="__next"]/main/div[1]/div/section/section/article[2]/ul').text.split('\n')
        try:
            dday=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[1]/div/section/section/article[4]/span').text
        except:
            dday=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[1]/div/section/section/article[3]/span').text

        address=driver.find_element(By.CSS_SELECTOR,'#__next > main > div.JobDetail_contentWrapper__DQDB6 > div > section > section > article.JobWorkPlace_JobWorkPlace__5fs5x > div > div > span').text

        time.sleep(0.5)

        driver.execute_script(f"window.scrollBy(0, {500});")
        time.sleep(1)
        more_inf = driver.find_element(By.XPATH,"/html/body/div[1]/main/div[1]/div/section/section/article[1]/div/button")
        more_inf.send_keys(Keys.ENTER)
        time.sleep(0.5)                                 

        position=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div[1]/div/section/section/article[1]').text.split('\n')[1:]

        time.sleep(0.5)
        tec_tag_list.append(tec_tag)
        tag_list.append(tag)
        dday_list.append(dday)
        address_list.append(address)
        position_list.append(position)

        time.sleep(0.5)
        print(idx)
        time.sleep(0.5)
    driver.quit()
```
