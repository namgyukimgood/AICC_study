2. Survived 컬럼에 들어가 있는 값을 쉬운 표현으로 바꿔주세요.

5. Embarked 컬럼과 Pclass 컬럼을 One Hot Encoding 해주세요

앞서 작성한 코드를 활용하여 Embarked 컬럼과 Pclass 컬럼을 충분히 분석하였습니다. 이번에는 이 두 개의 컬럼을 의사결정나무(Decision Tree)와 같은 머신러닝(Machine Learning) 알고리즘에 넣을 수 있도록 One Hot Encoding을 하도록 합니다. 먼저 Embarked 컬럼을 바탕으로, 다음의 세 개의 컬럼을 새롭게 만들고 싶습니다.

Embarked(C) - Embarked 컬럼 안에 있는 값이 C면 True, 아니면 False가 들어갑니다.
Embarked(S) - Embarked 컬럼 안에 있는 값이 S면 True, 아니면 False가 들어갑니다.
Embarked(Q) - Embarked 컬럼 안에 있는 값이 Q면 True, 아니면 False가 들어갑니다.
최종적으로는 다음의 결과가 나와야 합니다.

```python
train.head()
```

```python
train["Embarked(C)"] = train["Embarked"] == "C"
train["Embarked(S)"] = train["Embarked"] == "S"
train["Embarked(Q)"] = train["Embarked"] == "Q"

# Embakred를 기준으로 One Hot Encoding 컬럼을 만들고, 
# 관련된 컬럼들을 출력하여 비교합니다.
# 원핫 인코딩(One-Hot Encoding)은 범주형 데이터를 
# 숫자로 변환하는 방법 중 하나입니다. 주로 머신러닝 
# 알고리즘에서 범주형 데이터를 입력으로 사용할 때 
# 많이 적용됩니다. ? 잘 모르겠다 
```

```python
train[["Embarked", "Embarked(Status)", "Embarked(C)", "Embarked(S)", "Embarked(Q)"]].head(3)
```

Pclass 컬럼도 비슷하게 다음의 세 개의 컬럼이 만들어져야 합니다.

Pclass(F) - Pclass 컬럼 안에 있는 값이 1이면 True, 아니면 False가 들어갑니다.

8. 나이를 일정 구역(범위)으로 나눠서, 구역(범위)마다의 생존률을 보여주세요.

이번에는 나이(Age)별 생존률을 확인하고 싶습니다. 다만 나이 컬럼은 숫자이기 때문에, 그대로 쓰지 않고 일정 구역으로 나눈 뒤 생존률의 통계를 내는 것이 보기 편할 것입니다. 나이 컬럼을 다음의 세 구역으로 나눠봅니다.

Young : 나이가 15세 미만인 승객.
Medium : 나이가 15세 이상이고 30세 미만인 승객.
Old : 나이가 30세 이상인 승객.
최종적으로는 다음의 결과가 나와야 합니다.

```python
train.loc[train["Age"] < 15, "AgeType"] = "Young"
#pandas를 사용하여 train 데이터프레임에서 특정 조건을 
#만족하는 행에 대해 새로운 값을 할당하는 방법입니다.
#Agetype 이란 새로운 column이 생긴다 
```

```python
train.loc[(train["Age"] >= 15) & (train["Age"] < 30), "AgeType"] = "Medium"
```

```python
train.loc[train["Age"] >= 30, "AgeType"] = "Old"
```

```python
train[["Age", "AgeType"]].head(10)
```

이번엔, 위 조건에서 AgeType별 해당하는 승객의 평균 생존률을 구해봅니다. (예: 구역(범위)가 Young 인 경우, 약 57.7%가 나와야 합니다.)

```python
train.groupby("AgeType")["Survived"].mean().to_frame()
#train 데이터프레임에서 AgeType 열을 기준으로 그룹화하고, 
#각 그룹에 대해 Survived 열의 평균값을 
#계산한 다음, 그 결과를 데이터프레임 형식으로 반환합니다.
#데이터프레임으로 변환 (to_frame()): 계산된 평균값 
#시리즈를 데이터프레임으로 변환합니다.
```

10. Pclass별 나이(Age)의 평균을 구한 뒤 빈 값에 채워주세요.

이번에는 나이(Age) 컬럼의 빈 값을 채우려고 합니다. 일반적으로 가장 많이 하는 방식은 평균(mean)값을 구한 뒤, 이를 빈 값에 채워넣는 것인데, 이번에는 전체 평균 나이가 아닌, 각 객실 등급(Pclass)별 평균 연령을 구해서 해당 객실 등급에 빈 Age 값을 채우도록 하겠습니다.

객실 등급(Pclass)이 1등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이 값을 채워줍니다.
객실 등급(Pclass)이 2등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이 값을 채워줍니다.
객실 등급(Pclass)이 3등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이 값을 채워줍니다.

16. 호칭(Title) 컬럼을 정리해주세요.

호칭(Title) 컬럼은 생존자와 사망자를 구문하는데 중요한 역할을 담당합니다만, 너무 종류가 많은 단점이 있습니다.

가령 호칭(Title) 컬럼은 Mr, Mrs, Miss, Master가 전체 승객의 호칭의 대부분을 차지하며, Capt, Col, Don, Sir 등과 같은 컬럼들은 모수가 작아서 오히려 신뢰할 수 없습니다.

가령 Sir라는 호칭을 가진 승객의 생존률이 100%라고 하더라도, 전체 승객애서 오직 1명만이 Sir라는 호칭을 사용하고 있기 때문에, "Sir라는 호칭을 사용하는 사람은 언제나 생존률이 높다."고 이야기하기에는 어렵습니다.

```python
pd.crosstab(train["Title"], train["Survived"], margins=True)

# crosstab을 활용하여 각 호칭(Title)별 생존자의 
# 총 인원수와 사망자의 총 인원수를 출력합니다.

# pd.crosstab(train["Title"], train["Survived"], margins=True) 
# 명령어는 pandas에서 교차표
# (crosstabulation, contingency table)를 
# 생성하는 함수입니다. 
# 이 명령어는 train 데이터프레임에서 Title 열과
# Survived 열 간의 교차표를 만들어주며, 
# 추가적으로 행과 열의 합계를 포함한 결과를 반환합니다.

# 이 교차표는 각 호칭(Title)에 따른 생존자 수와
# 비생존자 수를 나타내며, 마지막 열과 행에는 각각 
# 열의 합계와 행의 합계가 포함되어 있습니다 
# (margins=True 설정으로 인해).
```

```python
train.groupby(["Title",'Survived'])['Survived'].agg(['count']).unstack()

# pandas를 사용하여 train 데이터프레임을 Title과 Survived 
# 열을 기준으로 그룹화한 후,
# 각 그룹에 대해 카운트(count)를 계산하고 이를 피벗하여 
# 표시하는 작업입니다.

# 여기서 Title과 Survived가 다중 인덱스로 설정되어 있습니다.

# 이 상태에서 unstack()을 적용하면, 
# 내부의 하위 인덱스(Survived)를 열로 옮기고, 
# Title을 행 인덱스로 사용하여 테이블을 재배열합니다.
```

```python
def get_title_status(title):
    title = title.strip()
    if title == "Mr" or title == "Miss" or title == "Mrs" or title == "Master":
        return title
    else: 
        return "Others"
#이번에는 get_cleaned_title 이름의 함수를 정의합니다.
# 이 함수에서는 Mr, Mrs, Miss, Master 가 
#아닌 호칭은 전부 Other라고 반환합니다.

# 모든 Name 컬럼 데이터에 get_cleaned_title 함수를 
#적용한 뒤 그 결과를 Title(Clean)이라는 이름의 새로운 
#컬럼에 대입합니다.
# 이름(Name) 컬럼과 Title 컬럼, 
#그리고 Title(Clean) 컬럼만을 출력합니다.
# 다만 이번에는 head가 아니라 tail을 
#활용해 마지막 다섯 컬럼만 출력합니다.
```

```python
train["Title(Status)"] = train["Title"].apply(get_title_status)
```

17. 호칭(Title) 컬럼으로 결혼 유무를 파악해주세요.

앞서 우리가 추출한 호칭 컬럼의 주요 값에 대한 설명은 다음과 같습니다.

Mr - 결혼한 남성을 의미합니다.
Master - 결혼하지 않은 남성을 의미합니다.
Mrs - 결혼한 여성을 의미합니다.
Miss - 결혼하지 않은 여성을 의미합니다.
그러므로 호칭(Title) 컬럼값에서는 추가로 결혼 여부를 뽑아낼 수 있습니다. 호칭이 Mr이거나 Mrs이면 결혼했다고 간주하고, Master이거나 Miss이면 결혼하지 않았다고 간주할 수 있습니다. (나머지 호칭은 NaN을 집어넣으면 좋겠습니다)

호칭(Title) 컬럼을 활용하여 결혼 유무(Married)를 추출해주세요, 최종적으로 다음의 결과가 나와야 합니다.

또한 결혼 유무(Married)별 생존률의 차이도 구해주세요. 최종적으로 다음의 결과가 나와야 합니다.
