```python
one = data[data['흡연상태'] == 1]
two = data[data['흡연상태'] == 2]
three = data[data['흡연상태'] == 3]
```

```python
one.head(3)
```

```python
plt.figure(figsize=(9, 3))
sns.kdeplot(one['허리둘레'])    #파랑
sns.kdeplot(two['허리둘레'])    #주황
sns.kdeplot(three['허리둘레'])  #녹색
plt.legend(labels=['흡연 안 함', '흡연 했지만 그만 둠', '흡연 중'])
```
