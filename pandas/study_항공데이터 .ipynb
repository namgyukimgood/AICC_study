```python
for a, row_a in airport_code.iterrows() :
    c = row_a['국  가']
    
    for j, row_c in country_code.iterrows() :
        if row_c['나라 이름'] == c :
            airport_code.loc[a,'영문 이름'] = row_c['alpha-3']

#iterrows() 메소드를 사용하여 각 행의 인덱스 a와 
#행의 데이터 row_a를 반환합니다.
#현재 행 row_a에서 국 가 열의 값을 변수 c에 저장합니다.
#iterrows() 메소드를 사용하여 각 행의 인덱스 j와 
#행의 데이터 row_c를 반환합니다.
#country_code 데이터프레임의 현재 행 row_c에서 
#나라 이름 열의 값이 변수 c와 같은지 확인합니다.
#airport_code 데이터프레임의 인덱스 a에 해당하는 
#행의 영문 이름 열에 row_c의 alpha-3 값을 할당합니다.

#airport_code 데이터프레임의 각 행에 대해 
#해당 국가의 3자리 알파벳 코드(alpha-3)를 
#country_code 데이터프레임에서 찾아 
#영문 이름 열에 업데이트합니다.
#이 코드는 airport_code 데이터프레임에 영문 이름이라는 
#새로운 열을 생성하고, 각 국가에 맞는 
#alpha-3 코드를 해당 열에 추가하는 역할을 합니다

#for문으로 구현 
# for a in range(len(airport_code)):  # airport_code 데이터프레임의 각 행을 인덱스를 사용하여 순회합니다.
#     row_a = airport_code.iloc[a]  # 현재 인덱스 a에 해당하는 행 데이터를 가져옵니다.
#     c = row_a['국  가']  # 현재 행의 '국  가' 값을 변수 c에 저장합니다.
    
#     for j in range(len(country_code)):  # country_code 데이터프레임의 각 행을 인덱스를 사용하여 순회합니다.
#         row_c = country_code.iloc[j]  # 현재 인덱스 j에 해당하는 행 데이터를 가져옵니다.
#         if row_c['나라 이름'] == c:  # country_code의 '나라 이름'이 c와 같은지 확인합니다.
#             airport_code.loc[a, '영문 이름'] = row_c['alpha-3']  # 조건이 참이면 airport_code의 현재 행에 '영문 이름' 열을 생성하고 값을 할당합니다.
```

```python
asia_depart = asia_.merge(country_code, 
                          left_on="출발 국가", 
                          right_on="alpha-3",
                          sort=False,how='left')

# asia_: 출발 국가와 도착 국가 열을 포함한 데이터프레임입니다.
# country_code: 국가 코드 정보를 포함한 데이터프레임입니다.
# merge(): 두 데이터프레임을 병합하는 pandas 메소드입니다.
# left_on="출발 국가": 왼쪽 데이터프레임(asia_)에서 
# '출발 국가' 열을 기준으로 사용합니다.
# right_on="alpha-3": 오른쪽 데이터프레임(country_code)에서
# 'alpha-3' 열을 기준으로 사용합니다.
# sort=False: 결과를 정렬하지 않습니다.
# how='left': 왼쪽 조인을 수행합니다. 
# 왼쪽 데이터프레임(asia_)의 모든 행을 포함하며, 
# 오른쪽 데이터프레임(country_code)에서 일치하는 
# 행이 있는 경우 연결합니다.

# 기준 열 설정: '출발 국가' 열과 'alpha-3' 열을 기준으로 
# 두 데이터프레임을 병합합니다.
# 왼쪽 조인: 왼쪽 데이터프레임(asia_)의 모든 행을 유지하며,
# 오른쪽 데이터프레임(country_code)에서 일치하는 
# 행이 있는 경우에만 연결합니다.
# 결과 생성: 병합된 결과를 새로운 데이터프레임인 
# asia_depart에 할당합니다.
```

```python
asia_dep_port = asia.groupby(["항공사","출발 공항","출발 국가","출발국가이름","도착 공항","도착 국가","년도","월"])
# asia 데이터프레임을 여러 열을 기준으로 
# 그룹화하는 작업입니다. 
# 그룹화된 결과는 asia_dep_port라는 
# 그룹화된 객체(GroupBy object)로 반환됩니다.
```

```python
asia_dep_port.head(1)
```

```python
a = asia_dep_port.agg({'월':'count',
                       '년도':'count',
                       '출발 국가':'count', 
                       '도착 국가':'count',
                       '정원(전체)':'sum'}) 

# asia_dep_port: 그룹화된 데이터프레임 객체입니다. 
# groupby() 메소드로 그룹화된 후에는 각 그룹에 대한 
#연산을 수행할 수 있는 상태입니다.
# agg(): 집계 함수를 적용하는 pandas 메소드입니다.
# {'월':'count', '년도':'count', '출발 국가':'count', '도착 국가':'count', '정원(전체)':'sum'}은 각 열에 적용할 집계 연산을 지정합니다.
# '월': 'count': '월' 열의 각 그룹별로 개수를 세어라.
# '년도': 'count': '년도' 열의 각 그룹별로 개수를 세어라.
# '출발 국가': 'count': '출발 국가' 열의 각 그룹별로 개수를 세어라.
# '도착 국가': 'count': '도착 국가' 열의 각 그룹별로 개수를 세어라.
# '정원(전체)': 'sum': '정원(전체)' 열의 각 그룹별로 합계를 계산해라.
#각 집계 함수가 적용된 결과는 
#새로운 데이터프레임 a에 저장됩니다.
```
